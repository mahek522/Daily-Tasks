```python
import matplotlib.pyplot as plt
import pandas as pd
data = {"Salary": [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 32000, 35000, 31000, 37000]}
df = pd.DataFrame(data)
df.head()
plt.plot(df['Salary'], color = 'green',marker = '*', linestyle = ':' , linewidth = 2)
plt.grid()
plt.show()
```

```python
hr_sal = df[df['Dept'] == 'HR']['Salary']
hr_sal
```

```python
it_sal = df[df['Dept'] == 'IT']['Salary']
it_sal
```

```python
sales_sal = df[df['Dept'] == 'Sales']['Salary']
sales_sal
```

```python
market_sal = df[df['Dept'] == 'Marketing']['Salary']
market_sal
```

```python
fin_sal = df[df['Dept'] == 'Finance']['Salary']
fin_sal
```

```python
plt.boxplot([hr_sal, sales_sal, it_sal, market_sal, fin_sal], tick_labels=['HR', 'Sales', 'IT', 'Marketing', 'Finance'])
plt.show()
```

```python
# Pie Chart
salary_by_dept = df.groupby("Dept")["Salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct = '%1.2f', colors = ['green', 'blue', 'yellow', 'orange', 'red'], shadow=True, explode =[0.1, 0, 0.1, 0.2, 0.1])
plt.axis("equal")
plt.show()
```

```python
#barplot
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
sales_mean = sum(sales_sal)/len(sales_sal)
fin_mean = sum(fin_sal)/len(fin_sal)
market_mean = sum(market_sal)/len(market_sal)
```

```python
plt.bar(['HR', 'IT', 'Sales','Finance', 'Marketing'], [hr_mean, it_mean, sales_mean, fin_mean, market_mean], color = ['green', 'blue', 'yellow', 'orange', 'red'])
plt.show()
```

```python
plt.scatter(df['Age'], df['Salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary, Experience")
plt.show()
```

```python
color = {'HR' : 'yellow', 'IT' : 'blue', 'Sales' : 'orange', 'Finance' : 'red', 'Marketing' : 'green'}

for Dept, color in color.items():
    df_dept = df[df['Dept'] == 'Dept']
    plt.scatter(df_dept['Age'], df_dept['Salary'], c = color, label = Dept)
```
