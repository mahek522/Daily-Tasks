```python
'''Task 11: Cumulative Sum Plot
Create a DataFrame with:
day
daily_sales
Compute cumulative sales.
 Plot cumulative sales using a line chart.'''
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data2 = {'day': [1, 2, 3, 4, 5, 6, 7], 'daily_sales': [150, 200, 250, 300, 350, 400, 450]}
df2 = pd.DataFrame(data2)
df2['cumulative_sales'] = df2['daily_sales'].cumsum()
```

```python
df2['cumulative_sales']
```

```python
plt.plot(df2['day'], df2['cumulative_sales'], marker='o')
plt.title('Cumulative Sales Over Time')
plt.xlabel('Day')
plt.ylabel('Cumulative Sales')
plt.show()
```
