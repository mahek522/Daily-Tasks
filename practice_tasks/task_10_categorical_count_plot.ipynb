```python
'''Task 10: Categorical Count Plot
Create a DataFrame with:
product_category
Count occurrences.
 Plot a bar chart showing product frequency.'''
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'product_category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Clothing', 'Books', 'Electronics', 'Clothing']
}
df = pd.DataFrame(data)
category_counts = df['product_category'].value_counts()
category_counts.plot(kind='bar')
plt.title('Product Frequency by Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```
