```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = {"Name" : ['Alex', 'Carry', 'John', 'James', 'Kevin', 'Tim', 'David', 'Neeraj', 'Sunil', 'Kelvin'], "Age" : [np.nan, 27, 30, np.nan, 29, 24, np.nan, 34, 35, np.nan]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Age'] = df['Age'].fillna(df["Age"].mean())
```

```python
df
```

```python
plt.bar(df['Name'], df['Age'], color = 'green', align = 'center')
plt.show()
```
