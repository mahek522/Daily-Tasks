```python
'''Task 3: GroupBy + Bar Chart
Create a DataFrame with:
department
salary
Calculate average salary per department.
 Visualize using a bar chart.'''

import matplotlib.pyplot as plt
import pandas as pd
data = {'Sales' : [24000, 25000, 27000, 37000, 30000, 34000, 40000, 23000, 28000, 36000, 35000, 50000]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df['Dept'] = ['HR', 'IT', 'Finance', 'Sales', 'Marketing', 'IT'] * 2
```

```python
df
```

```python
df = df.rename(columns={'Sales' : 'Salary'})
```

```python
df
```

```python
hr_sal = df[df['Dept'] == 'HR']['Salary']
hr_sal
```

```python
it_sal = df[df['Dept'] == 'IT']['Salary']
it_sal
```

```python
sales_sal = df[df['Dept'] == 'Sales']['Salary']
```

```python
finance_sal = df[df['Dept'] == 'Finance']['Salary']
```

```python
marketing_sal = df[df['Dept'] == 'Marketing']['Salary']
```

```python
salary_by_dept = df.groupby("Dept")["Salary"].sum()
salary_by_dept
```

```python
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
sales_mean = sum(sales_sal)/len(sales_sal)
finance_mean = sum(finance_sal)/len(finance_sal)
marketing_mean = sum(marketing_sal)/len(marketing_sal)
```

```python
plt.bar(['HR', 'IT', 'Sales','Finance', 'Marketing'], [hr_mean, it_mean, sales_mean, finance_mean, marketing_mean], color = ['green', 'blue', 'yellow', 'orange', 'red'])
plt.show()
```
