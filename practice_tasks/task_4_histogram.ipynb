```python
'''Task 4: Histogram
Create a DataFrame with:
marks (50 random values between 0 and 100)
Plot a histogram with appropriate bins.
'''

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
marks = np.random.randint(0,100,50)
```

```python
marks
```

```python
df = pd.DataFrame(marks)
```

```python
df
```

```python
df.name = "Marks"
```

```python
df
```

```python
plt.hist(df, bins = 5)
plt.xlabel("Marks")
plt.show()
```
