```python
'''Task 8: Multi-Column Plot
Create a DataFrame with:
month
revenue
expenses
Plot both revenue and expenses on the same line graph.
'''
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'revenue': [2000, 3000, 4000, 3500, 5000, 6000],
    'expenses': [1500, 2500, 3000, 2800, 4000, 4500]
}
```

```python
df = pd.DataFrame(data)
plt.plot(df['month'], df['revenue'], label='Revenue', marker='o')
plt.plot(df['month'], df['expenses'], label='Expenses', marker='o')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Revenue vs Expenses Over Months')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
