```python
'''Task 7: Time Series Plot
Create a DataFrame with:
date (last 10 days)
temperature
Convert date to datetime.
 Plot a time-series line graph.
'''
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Create date range for the last 10 days
date_range = [datetime.now() - timedelta(days=i) for i in range(10)][::-1]
# Sample temperature data
temperature_data = [22, 21, 23, 24, 22, 25, 26, 24, 23, 22]
# Create DataFrame
data = {'date': date_range, 'temperature': temperature_data}
```

```python
df = pd.DataFrame(data)
# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# Plot time-series line graph
plt.plot(df['date'], df['temperature'])
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')
plt.title('Temperature Over the Last 10 Days')
plt.xticks(rotation=45)
plt.show()
```
