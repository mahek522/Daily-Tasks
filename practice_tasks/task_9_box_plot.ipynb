```python
'''Task 9: Box Plot
Create a DataFrame with:
class
marks
Plot a box plot to compare mark distribution across classes.'''
import matplotlib.pyplot as plt
import pandas as pd
data_box = {'class': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
            'marks': [85, 90, 78, 88, 92, 80, 75, 70, 95]}
df_box = pd.DataFrame(data_box)
df_box.boxplot(column='marks', by='class')
plt.title('Box Plot of Marks by Class')
plt.xlabel('Class')
plt.ylabel('Marks')
plt.show()
```
