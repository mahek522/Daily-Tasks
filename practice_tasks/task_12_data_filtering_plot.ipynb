```python
'''Task 12: Data Filtering + Plot
Create a DataFrame with:
city
population
Filter cities with population greater than 1 million.
 Plot a bar chart for the filtered cities.'''
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'City': ['CityA', 'CityB', 'CityC', 'CityD', 'CityE'],
        'Population': [500000, 1500000, 2000000, 800000, 1200000]}
df = pd.DataFrame(data)
filtered_df = df[df['Population'] > 1000000]
filtered_df
```

```python
plt.bar(filtered_df['City'], filtered_df['Population'])
plt.title('Cities with Population > 1 Million')
plt.xlabel('Cities')
plt.ylabel('Population')
plt.show()
```
